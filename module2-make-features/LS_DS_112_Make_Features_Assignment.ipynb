- Convert the `term` column from string to integer.
- Make a column named `loan_status_is_great`. It should contain the integer 1 if `loan_status` is "Current" or "Fully Paid." Else it should contain the integer 0.
- Make `last_pymnt_d_month` and `last_pymnt_d_year` columns.

```python
# define a function
def stripmonths(input):
  return float(input.strip().strip('months'))
```

```python
# test said function
test="36 months	"
test2=stripmonths(test)
print(test)
print(test2)
```

```python
# apply said function to the entire column
dfclean["term"]=dfclean['term'].apply(stripmonths)
```

```python
# inspect said data
dfclean.head()
```

```python
# - Make a column named `loan_status_is_great`. 
# It should contain the integer 1 if `loan_status` is "Current" or "Fully Paid." 
# Else it should contain the integer 0.
# lets make another function
def verifyloan(loan):
  if loan=='Current':
    return 1
  elif loan=='Fully Paid':
    return 1
  else:
    return 0
```

```python
dfclean['loan_status_is_great']=dfclean['loan_status'].apply(verifyloan)
```
